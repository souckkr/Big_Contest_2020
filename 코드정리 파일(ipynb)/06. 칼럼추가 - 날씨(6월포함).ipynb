# 1. 2019년 실적데이터 전처리

## 1.1 데이터 로드

```python
# data = pd.read_excel('data/2020 빅콘테스트 데이터분석분야-챔피언리그_2019년 실적데이터_v1_200818.xlsx', header=1)
data = pd.read_csv('data/train_data_05.csv', index_col = 0)
data.head()
```

```python
# 2019 기상데이터
w19 = pd.read_csv('data/2019날씨.csv', encoding = 'cp949')
w19.head()
```

## 1.2 데이터 타입 확인

```python
data.info()
```

```python
w19.info()
```

## 1.3 칼럼 추가

### 1.3.1 기상데이터 칼럼 추가

```python
data2 = data.copy()
w19_2 = w19.copy()
```

```python
# 서울-수도권 기상 정보 평균
w19_2 = w19_2.groupby('일시').mean()
w19_2.head()
```

```python
del w19_2['지점']
```

```python
w19_2 = w19_2.rename(columns = {'기온(°C)' : '평균기온(°C)'})
```

```python
w19_2 = w19_2.fillna(0)
```

```python
w19_2 = w19_2.reset_index()
```

```python
# 20분 단위인 실적데이터와 병합하기 위해 시간단위 기준칼럼 새로 생성
w19_2['일시_시'] = pd.to_datetime(w19_2['일시'], format = '%Y-%m-%d %H')

w19_2['일시_시'] = w19_2['일시_시'].dt.strftime('%Y-%m-%d %H')
```

```python
# 실적데이터도 병합을 위한 기준칼럼 생성

data2 = data2.reset_index()

data2['일시_시'] = pd.to_datetime(data2['방송일시'], format = '%Y-%m-%d %H')

data2['일시_시'] = data2['일시_시'].dt.strftime('%Y-%m-%d %H')
```

```python
# merge
data3 = pd.merge(data2,w19_2, on = '일시_시')
```

```python
# del data3['index']
del data3['일시_시']
del data3['일시']
```

```python
data3.head()
```

### 1.3.2 체감온도 칼럼 추가
  
- 체감온도(°C)
    - 13.12+0.6215×T-11.37V0.16+0.3965V0.16×T   
    - T = 기온, V = 풍속(km/h)  

```python
# 계산을 위해 풍속(km/h)으로 고치기

data3['풍속(km/h)'] = data3['풍속(m/s)'] *3.6

data3['체감온도'] = round(13.12+0.6215*data3['평균기온(°C)']-11.37*data3['풍속(km/h)']**0.16+0.3965*data3['풍속(km/h)']**0.16*data3['평균기온(°C)'],1)
```

```python
del data3['풍속(km/h)']
```

```python
data3[['방송일시','체감온도']]
```

### 1.3.3 강수여부 칼럼 추가


- 비옴  
    - 강수량 >= 10mm

```python
data3.loc[data3['강수량(mm)'] >= 1.0, '비옴'] = 1
data3.loc[data3['강수량(mm)'] < 1.0, '비옴'] = 0

# float -> int
data3['비옴'] = data3['비옴'].astype(int)
```

```python
data3.head()
```

```python
# 파일 저장
data3.to_csv('data/train_data_06.csv', index=False)
```
