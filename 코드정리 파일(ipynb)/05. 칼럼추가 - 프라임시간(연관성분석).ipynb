# 연관성분석(장바구니분석)

```python
te = TransactionEncoder()
te_ary = te.fit(df_val).transform(df_val)
dataset = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
dataset
```

```python
frequent_itemsets = apriori(dataset, min_support=0.001, use_colnames=True) #총 1672개 조합
```

```python
combin = []
for i in range(len(frequent_itemsets)):
    if len(frequent_itemsets.iloc[i][1]) == 3:
        combin.append(sorted(list(frequent_itemsets.iloc[i][1])))
```

# 시청률로 연관조합 필터링

```python
amount = data.groupby('상품군')['취급액','시청률'].mean().reset_index()
amount
```

```python
data['상품군평균시청률'] = np.nan
```

```python
data['상품군'][0]
```

```python
for i in range(len(data)):
    for j in range(len(amount)):
        if data['상품군'][i] == amount['상품군'][j]:
            data['상품군평균시청률'][i] = amount['시청률'][j]
    
```

```python
data.info()
```

```python
tv = data.groupby(['방송시간','방송요일','상품군','상품군평균시청률'])['시청률'].mean().reset_index()
tv
```

```python
# 조합중 상품평균시청률보다 낮은것 제거
new_combin = []
for i in range(len(tv)):
    for j in combin:
        if tv.iloc[i][3] <= tv.iloc[i][4]:
            if tv['방송시간'][i] == j[0] and tv['방송요일'][i] == j[1] and tv['상품군'][i] == j[2]:
                new_combin.append(j)
```

```python
new_combin
```

```python
len(new_combin)
```

```python
newcombindata = pd.DataFrame(new_combin, columns=['방송시간','방송요일','상품군'])
```

```python
newcombindata['상품군'].value_counts()
```

# Train 데이터에 적용

```python
data['프라임시간'] = np.nan
```

```python
# 최적의 조합(프라임시간)이라면 1을 삽입
for i in range(len(data)):
    for j in new_combin:
        if data['방송시간'][i] == j[0] and data['방송요일'][i] == j[1] and data['상품군'][i] == j[2]:
            data['프라임시간'][i] = 1
```

```python
# 최적의 조합(프라임시간)이 아니라면 0을 삽입
data['프라임시간'] = data['프라임시간'].fillna(0)
```
