# 시계열분석

```python
# pip install --upgrade pip
```

```python
from fbprophet import Prophet

cloth = new_time[new_time['상품군']=='의류']
cloth.drop('상품군',axis=1,inplace=True)
cloth.columns = ['ds','y']
cloth
```

```python
#   m = Prophet(mcmc_samples=300,changepoint_prior_scale=0.1) # Default growth='linear'
m = Prophet()
m.add_seasonality(name='monthly', period=30.5, fourier_order=5)
m.fit(cloth)
```

```python
future = m.make_future_dataframe(periods=30)
future.tail()
```

```python
forecast = m.predict(future)
forecast
```

```python
# forecast.set_index('ds',inplace=True)
forecast.loc['2020-07-01':'2020-07-07']
```

```python
from fbprophet.plot import add_changepoints_to_plot

fig = m.plot(forecast)
a = add_changepoints_to_plot(fig.gca(), m, forecast)
fig2 = m.plot_components(forecast)
plt.show()
```

## 상품군 별 시계열분석

```python
from fbprophet import Prophet

tv_rate = []
for i in new_time['상품군'].unique():
    product_data = new_time[new_time['상품군']==i]
    product_data.drop('상품군',axis=1,inplace=True)
    product_data.columns = ['ds','y']
    m = Prophet(yearly_seasonality=True, daily_seasonality=True) # Default growth='linear'
    m.fit(product_data)
    future = m.make_future_dataframe(periods=30)
    forecast = m.predict(future)
    
    fig = m.plot(forecast)
    a = add_changepoints_to_plot(fig.gca(), m, forecast)
    plt.title(i + ' 편성시간 추이')
    fig2 = m.plot_components(forecast)
    plt.show()
    
    forecast.set_index('ds',inplace=True)
    last = forecast.loc['2020-06-03':'2020-06-23','trend'].mean()
    pred = forecast.loc['2020-07-01':'2020-07-07', 'trend'].mean()

    compare = round(((pred-last)/abs(last)) * 100, 2)
    tv_rate.append([i,compare])
    print(f'3주간 평균 트렌드 : {last}, 7월 1주차 예상 트렌드평균 : {pred}, 트렌드증감률 : {compare}')
    print('='*100)
```

```python
tv_rates = pd.DataFrame(tv_rate, columns=['상품군','편성시간증감률'])
tv_rates
```

```python
june['소분류']=june['소분류'].fillna('무형')
```

```python
data_june = june.groupby(['방송일시','상품군']).agg({'소분류':'max','취급액':'sum'}).reset_index()
data_june
```

```python
data_june['방송일시'] = pd.to_datetime(data_june['방송일시'])
data_june['방송일'] = data_june['방송일시'].dt.date
data_june['방송시'] = data_june['방송일시'].dt.hour
data_june['방송시간'] = data_june['방송일시'].dt.time
data_june['방송요일'] = data_june['방송일시'].dt.day_name()
data_june.set_index('방송일시',inplace=True)
```

```python
data_june_3w = data_june.loc['2020-06-03 06:20:00':'2020-06-24 01:40:00']
data_june_3w = data_june_3w.reset_index()
data_june_3w
```

```python
data_june_3w.drop('방송일시',axis=1, inplace=True)
```

```python
data_june_3w
```

```python
data_june_info = data_june_3w.groupby(['상품군']).agg({'취급액':'mean'}).reset_index()
data_june_info.columns = ['상품군','상품군별평균취급액']
data_june_info
```

```python
data_june_merge = pd.merge(data_june_3w,data_june_info,how='left',on='상품군')
data_june_merge
```

```python
data_june_merge.info()
```

```python
data_june_merge_max = data_june_merge.groupby(['방송요일','방송시간','소분류','상품군','상품군별평균취급액'])['취급액'].max().reset_index()
```

## 6월 프라임 조합 이용하여 7월 1주차 편성표구성

```python
# !pip install mlxtend
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data_june_merge_max = data_june_merge.groupby(['방송요일','방송시','소분류','상품군','상품군별평균취급액'])['취급액'].max().reset_index()
data_june_merge_max
```

```python
data_june_merge_max["방송시간"] = data_june_merge_max["방송시"].map(str) + "시"
data_june_merge_max["기준"] = data_june_merge_max["방송요일"].map(str) + data_june_merge_max["방송시간"].map(str) +data_june_merge_max["상품군"].map(str)
data_june_merge_max
```

```python
# 취급액 필터링
new_june=data_june_merge_max[data_june_merge_max['상품군별평균취급액']<=data_june_merge_max['취급액']]
```

```python
df = new_june[['방송요일','방송시간','상품군']]
df
```

```python
df_val = df.values  #array로 가능(Dataframe불가)
```

```python
df_val
```

```python
te = TransactionEncoder()
te_ary = te.fit(df_val).transform(df_val)
dataset = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
frequent_itemsets = apriori(dataset, min_support=0.005, use_colnames=True)
frequent_itemsets
```

```python
combin = []
for i in range(len(frequent_itemsets)):
    if len(frequent_itemsets.iloc[i][1]) == 3:
        combin.append(sorted(list(frequent_itemsets.iloc[i][1])))
```

```python
len(combin)
```

```python
combindata = pd.DataFrame(combin, columns=['방송시간','방송요일','상품군'])
combindata["기준"] = combindata["방송요일"].map(str) + combindata["방송시간"].map(str) +combindata["상품군"].map(str)
combindata
```

```python
data_june_merge_max2 = data_june_merge_max[['기준','방송시','소분류']]
data_june_merge_max2
```

```python
new_combin = pd.merge(combindata,data_june_merge_max2,on='기준',how='left')
new_combin.drop(['기준','방송시간'], axis=1, inplace=True)
new_combin
```

```python
punsungs = pd.pivot_table(new_combin, index = '방송요일', columns='방송시', values = '상품군',aggfunc='first')
punsungs
```

```python
punsungs.to_excel('data/punsungs.xlsx')
```
