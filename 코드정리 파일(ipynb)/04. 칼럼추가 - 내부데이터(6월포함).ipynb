# 1. 2019년 실적데이터 전처리

## 1.1 데이터 로드

```python
data = pd.read_csv('data/train_data_03.csv', index_col = 0)
data.head()
```

## 1.2 데이터 타입 확인

```python
data.info()
```

```python
# 데이터 타입 변환
data['방송일시'] = pd.to_datetime(data['방송일시'], format='%Y-%m-%d %H:%M:%S')
data['마더코드'] = data['마더코드'].astype(object)
data['상품코드'] = data['상품코드'].astype(object)
```

## 1.3 결측치 처리

### 1.3.1 노출(분) 결측치 처리
- 방송일시가 동일한 상품은 노출 시간 또한 동일하다고 판단하여 같은 값으로 처리함.

```python
data2 = data.copy()
for i in range(1, len(data2)):
    if data2['방송일시'][i-1] == data2['방송일시'][i]:
        data2['노출(분)'][i] = data2['노출(분)'][i-1]
data2.head()
```

### 1.3.2 취급액 결측치 처리
- 판매단가가 0원인 무형, 보험 상품과 주문량이 0인 칼럼의 취급액은 NaN 값으로 표시되므로, 해당 칼럼을 제외함.

```python
data2 = data2.dropna(subset=['취급액']).reset_index(drop=True)
data2.isnull().sum()
```

## 1.4 칼럼 추가

### 1.4.1 주문량(건수) 칼럼 추가
- 건수 = 취급액/판매단가 (소수점 올림)

```python
data3 = data2.copy()
data3['건수'] = round(data3['취급액']/data3['판매단가'])
```

### 1.4.2 방송날짜, 방송월, 방송일, 방송요일, 방송시간, 평일/주말 칼럼 추가
- 방송요일 - 0:월, 1:화, 2:수, 3:목, 4:금, 5:토, 6:일
- 평일/주말 - 0:평일, 1:주말

```python
data3['방송날짜'] = data3['방송일시'].dt.date
data3['방송월'] = data3['방송일시'].dt.month
data3['방송일'] = data3['방송일시'].dt.day
data3['방송요일'] = data3['방송일시'].apply(lambda x:x.weekday())
data3['방송시간'] = data3['방송일시'].dt.time
data3['평일/주말'] = 0
data3.loc[(data3['방송요일']==5) | (data3['방송요일']==6), '평일/주말'] = 1
```

### 1.4.3 공휴일 칼럼 추가
- 0:공휴일 X, 1:공휴일 O

```python
# 공휴일이 담긴 파일 불러오기
holiday = pd.read_excel('data/국가공휴일.xlsx')
holiday.head()
```

```python
holiday.info()
```

```python
holiday['년'] = holiday['년'].astype(str)
holiday['월'] = holiday['월'].astype(str)
holiday['일'] = holiday['일'].astype(str)
holiday['방송날짜'] = holiday['년'] + '-' + holiday['월'] + '-' + holiday['일']
holiday['방송날짜'] = pd.to_datetime(holiday['방송날짜'])
holiday['공휴일'] = 1
holiday = holiday[['방송날짜', '공휴일']]
holiday.head()
```

```python
holiday['방송날짜'] = holiday['방송날짜'].astype(str)
data3['방송날짜'] = data3['방송날짜'].astype(str)
data3 = pd.merge(data3, holiday, on='방송날짜', how='left')
data3.head()
```

```python
# 공휴일이 아닌 값을 0으로 채우기
data3['공휴일'] = data3['공휴일'].fillna(0)
data3['공휴일'] = data3['공휴일'].astype(int)
data3['공휴일'].value_counts()
```

### 1.4.4 계절 칼럼 추가
- 봄:3 ~ 5월, 여름:6 ~ 8월, 가을:9 ~ 11월, 겨울:12 ~ 2월

```python
data3['계절'] = np.nan
data3.loc[(data3['방송월'] >= 3) & (data3['방송월'] <= 5), '계절'] = '봄'
data3.loc[(data3['방송월'] >= 6) & (data3['방송월'] <= 8), '계절'] = '여름'
data3.loc[(data3['방송월'] >= 9) & (data3['방송월'] <= 11), '계절'] = '가을'
data3.loc[(data3['방송월'] >= 12) | (data3['방송월'] <= 2), '계절'] = '겨울'
data3.head()
```

```python
data3['계절'].value_counts()
```

### 1.4.5 방송회차 칼럼 추가

```python
# 방송일시, 노출(분) 기준으로 정렬
data4 = data3.sort_values(by=['방송일시','노출(분)'])
data4 = data4.reset_index(drop=True)
```

```python
# 방송일시가 중복되는 행을 제외한 데이터셋 생성
drop_idx = []
for i in range(1, len(data4)):
    if data4['방송일시'][i-1] == data4['방송일시'][i]:
        drop_idx.append(i)
```

```python
test_data = data4.copy()
test_data = test_data.drop(drop_idx, axis=0)
test_data.head()
```

```python
test_data['방송회차'] = 0
cnt = 0
for idx1, idx2 in zip(test_data.index[:-1], test_data.index[1:]):
    cnt += 1
    if test_data.loc[idx1, '상품명'] == test_data.loc[idx2, '상품명']:
        test_data.loc[idx1, '방송회차'] = cnt
    else:
        test_data.loc[idx1, '방송회차'] = cnt
        cnt = 0

# 마지막 index 처리
idx1 = test_data.index[-1]
idx2 = test_data.index[-2]
if test_data.loc[idx1, '상품명'] == test_data.loc[idx2, '상품명']:
    cnt += 1
    test_data.loc[idx1, '방송회차'] = cnt
else:
    test_data.loc[idx1, '방송회차'] = 0
```

```python
test_data.head()
```

```python
test_data['방송회차'].value_counts()
```

```python
test_data2 = test_data[['방송회차']]
data5 = pd.merge(data4, test_data2, left_index=True, right_index=True, how='left')
data5.head()
```

```python
# 앞의 값으로 결측치 채워주기
data5['방송회차'] = data5['방송회차'].fillna(method = 'ffill')
data5['방송회차'] = data5['방송회차'].astype(int)
data5.head()
```

### 1.4.6 무이자/일시불 칼럼 추가
- 상품군별로 모델을 적용할 때 활용할 예정임.
- 독립표본 t-검정을 적용한 결과, t값이 10.582, p-value가 0으로 신뢰수준 99%하에서 무이자/일시불에 따른 전체 취급액에 차이가 있다고 볼 수 있음.
- 의류, 농수축에는 '무이자','일시불' 값이 존재하지 않으며, 가전, 침구는 차이가 없는 것으로 나타남.
- 신뢰수준 99%하에서 속옷, 주방, 이미용, 생활용품, 건강기능, 잡화, 가구의 무이자/일시불에 따른 취급액에 차이가 있는 것으로 보임.

```python
data5['무이자/일시불'] = np.nan
for i in range(len(data5)):
    if ('무이자' in data5['상품명'][i]) | ('(무)' in data5['상품명'][i]):
        data5['무이자/일시불'][i] = '무이자'
    elif ('일시불' in data5['상품명'][i]) | ('(일)' in data5['상품명'][i]):
        data5['무이자/일시불'][i] = '일시불'
```

```python
data5['무이자/일시불'].value_counts()
```

```python
lump_sum = data5[data5['무이자/일시불']=='일시불']
no_interest = data5[data5['무이자/일시불']=='무이자']

# Levene의 등분산 검정
l_result = stats.levene(lump_sum['취급액'], no_interest['취급액'])
print('LeveneResult : %.3f \np-value : %.3f' % (l_result))
```

```python
# 두 표본의 분산이 동일하지 않으므로, 등분산이 아닌 독립표본 t-검정 실행
result = stats.ttest_ind(lump_sum['취급액'], no_interest['취급액'], equal_var=False)
print('t statistic : %.3f \np-value : %.3f' % (result))
```

```python
# 상품군별 t-검정 실행
for i in data5['상품군'].unique():
#     l_result = stats.levene(lump_sum[lump_sum['상품군']==i]['취급액'], no_interest[no_interest['상품군']==i]['취급액'])
#     print(i, 'LeveneResult : %.3f \np-value : %.3f' % (l_result))
    result = stats.ttest_ind(lump_sum[lump_sum['상품군']==i]['취급액'], no_interest[no_interest['상품군']==i]['취급액'], equal_var=False)
    print(i, 't statistic : %.3f \np-value : %.10f' % (result))
```

```python
data5.head()
```

```python
# 파일 저장
data5.to_csv('data/train_data_04.csv', index=False)
```

### 2.4.3 계절 칼럼 추가
- 봄:3 ~ 5월, 여름:6 ~ 8월, 가을:9 ~ 11월, 겨울:12 ~ 2월

```python
data3['계절'] = np.nan
data3.loc[(data3['방송월'] >= 3) & (data3['방송월'] <= 5), '계절'] = '봄'
data3.loc[(data3['방송월'] >= 6) & (data3['방송월'] <= 8), '계절'] = '여름'
data3.loc[(data3['방송월'] >= 9) & (data3['방송월'] <= 11), '계절'] = '가을'
data3.loc[(data3['방송월'] >= 12) | (data3['방송월'] <= 2), '계절'] = '겨울'
data3.head()
```

### 2.4.4 방송회차 칼럼 추가

```python
# 방송일시, 노출(분) 기준으로 정렬
data4 = data3.sort_values(by=['방송일시','노출(분)'])
data4 = data4.reset_index(drop=True)
```

```python
# 방송일시가 중복되는 행을 제외한 데이터셋 생성
drop_idx = []
for i in range(1, len(data4)):
    if data4['방송일시'][i-1] == data4['방송일시'][i]:
        drop_idx.append(i)
```

```python
test_data = data4.copy()
test_data = test_data.drop(drop_idx, axis=0)
test_data.head()
```

```python
test_data['방송회차'] = 0
cnt = 0
for idx1, idx2 in zip(test_data.index[:-1], test_data.index[1:]):
    cnt += 1
    if test_data.loc[idx1, '상품명'] == test_data.loc[idx2, '상품명']:
        test_data.loc[idx1, '방송회차'] = cnt
    else:
        test_data.loc[idx1, '방송회차'] = cnt
        cnt = 0

# 마지막 index 처리
idx1 = test_data.index[-1]
idx2 = test_data.index[-2]
if test_data.loc[idx1, '상품명'] == test_data.loc[idx2, '상품명']:
    cnt += 1
    test_data.loc[idx1, '방송회차'] = cnt
else:
    test_data.loc[idx1, '방송회차'] = 0
```

```python
test_data.head()
```

```python
test_data['방송회차'].value_counts()
```

```python
test_data2 = test_data[['방송회차']]
data5 = pd.merge(data4, test_data2, left_index=True, right_index=True, how='left')
data5.head()
```

```python
# 앞의 값으로 결측치 채워주기
data5['방송회차'] = data5['방송회차'].fillna(method = 'ffill')
data5['방송회차'] = data5['방송회차'].astype(int)
data5.head()
```

### 2.4.5 무이자/일시불 칼럼 추가

```python
data5['무이자/일시불'] = np.nan
for i in range(len(data5)):
    if ('무이자' in data5['상품명'][i]) | ('(무)' in data5['상품명'][i]):
        data5['무이자/일시불'][i] = '무이자'
    elif ('일시불' in data5['상품명'][i]) | ('(일)' in data5['상품명'][i]):
        data5['무이자/일시불'][i] = '일시불'
```
