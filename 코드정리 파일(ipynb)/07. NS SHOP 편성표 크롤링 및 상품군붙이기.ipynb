```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from urllib import parse
from datetime import datetime
from selenium.webdriver import Chrome
from selenium.webdriver import ChromeOptions
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import re
import time
```

```python
#날짜 차이
from datetime import date

d0 = date(2020, 8 ,17)
d1 = date(2019, 12, 31)
delta = d0 - d1
delta.days  
```

```python
options = ChromeOptions()
options.add_argument('headless')

driver = Chrome()
driver.get(url)
result = []
```

```python
# 홈쇼핑 -> ns shop+ 로 이동
driver.find_element_by_xpath('//*[@id="8"]').click()   
driver.find_element_by_xpath('//*[@id="117"]').click()   
```

```python
import time, datetime
page = 0
# 2020-01-01 까지 날짜, 시간, 상품명 크롤링
while page != 230:
    soup = BeautifulSoup(driver.page_source)
    day = soup.select_one('span.date').text.strip()[:-5]
    day = '-'.join(re.findall("\d+", day))
    print(day)
    item_list = soup.select('div.tblType.list table tbody tr')
    error_cnt = 0
    
    for item in item_list:
        hour = item.select_one('td.txtC').text.strip()
        program = item.select_one('td.txtL').text.split('\n')[0]
        result.append([day,hour,program])
    driver.find_element_by_class_name('prev').click()
    page += 1
    time.sleep(2)
    
```

```python
df = pd.DataFrame(result, 
                  columns=['날짜','시간', '상품명'])
df
```

```python
# 날짜, 시간으로 오름차순 정렬
df_rev = df.sort_values(by=['날짜','시간'], axis=0).reset_index()
df_rev
```

```python
df_rev = df_rev.drop('index', axis=1)
```
